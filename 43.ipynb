```python
import sys
import warnings
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

warnings.filterwarnings('ignore')
np.set_printoptions(threshold=sys.maxsize)
```

```python
### Problem 4.3
file = '4/SMAI-Dataset-problem-4.3/colors.pkl'
data = np.load(file, allow_pickle=True)

n = 7
gmm = GaussianMixture(n_components=n)
gmm.fit(data)
means = gmm.means_
covariances = gmm.covariances_
labels = gmm.predict(data)

plt.figure(figsize=(10, 6))
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', s=20)
plt.colorbar()
plt.title('Scatter Plot of Data Points')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()
```

```python
# print(means)
print(covariances)
print(covariances.round())
```

```python
def generate_sample_dataset(n, means, covariances, num_samples=1000):
    samples = []
    for i in range(n):
        component_samples = np.random.multivariate_normal(means[i], covariances[i], num_samples)
        samples.append(component_samples)
    return np.vstack(samples)

sample_data = generate_sample_dataset(n, means, covariances, num_samples=1000)
print(sample_data)
```
